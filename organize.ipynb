```python
import os

# Iterate through the data folder and print all the file names
for root, dirs, files in os.walk("data"):
    for file in files:
        # Create the folder
        folder_name = os.path.splitext(file)[0]
        os.mkdir("data/" + folder_name)
        path = "data/" + folder_name

        # Create the performance folder
        os.mkdir(os.path.join(path, "performance"))
        perf_path = os.path.join(path, "performance")

        # Move the file to the performance folder
        source_path = os.path.join(root, file)
        dest_path = os.path.join(perf_path, file)
        os.rename(source_path, dest_path)
```
