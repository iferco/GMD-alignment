```python
import pandas as pd
```

We read the full evaluation dataset

```python
df = pd.read_csv("/content/GMD-alignment/evaluation/full_evaluation.csv")
```

In order to compare the evaluation metrics with other features, we compute the diference between time duration and tempo extracted on our .mid and .wav audio for each muiñeira. The value is rounded to the nearest multiple of 10.

```python
df["difDuration"] = df["durationWav"]-df["durationMidi"]
df['rounded_difDuration'] = round(df['difDuration'] / 10) * 10

df["difTempo"] = df["tempoWAV"]-df["tempoMidi"]
df['rounded_difTempo'] = round(df['difTempo'] / 10) * 10
```

```python
df.to_csv("/content/GMD-alignment/final_dataset.csv", index = False)
```

## More information

For further information these plots can be observed

On this first figure, for each muiñeira the *coef_corr_chroma* is represented with an empty mark and the *coeff_corr_spectra* with a filled mark. Then the darkest colors are the one with greater difference on duration of audio. Finally the squares represent audios from youtube and the circles audio recordings.

```python
import altair as alt
alt.Chart(df).mark_point(filled=True, size = 120).encode(
    x='txt:N',
    y='coef_corr_chroma:Q',
    shape = 'audio_kind:N',
    color = 'rounded_difDuration:O'
)+ alt.Chart(df).mark_point(color="red",filled=False, size=120).encode(
    x='txt:N',
    y='coef_corr_spectra:Q',
    shape = 'audio_kind:N',
    color = 'rounded_difDuration:O'
)
```

On this second figure, for each muiñeira the *distance_chroma* is represented on blue and the *distance_spectra* in red. Then the a different shape of the mark represent the provice form origin of the muiñeira. Finally the bigger marks come from youtube and the smaller one come from recording files.

```python
import altair as alt
alt.Chart(df).mark_point().encode(
    x='txt:N',
    y='distance_chroma:Q',
    size = 'audio_kind:N',
    shape = 'province:N'
)+ alt.Chart(df).mark_point(color="red").encode(
    x='txt:N',
    y='distance_spectra:Q',
    size = 'audio_kind:N',
    shape = 'province:N'
)
```
