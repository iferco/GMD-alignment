We read the data from the directory using os.walk() to extract the paths for all the Muiñeiras dataset. 
We only keep the pathsthat contain the word 'alignments' in them, the alighments are what we want to evaluate after the algorithm DTW has runned.

```python
data_dir = '/content/GMD-alignment/data'
list_muiñeiras = []
for root, dirs, files in os.walk(data_dir):
    for dir in dirs:
        #dir - name of the folder; path - path to the folder
        path=os.path.join(root,dir)
        if 'score' not in path and 'performance'not in path:
            list_muiñeiras.append(path)
list_muiñ_alignments = []

for path in list_muiñeiras:
  if 'alignments' in path:
    list_muiñ_alignments.append(path)
    #print(path)
print(list_muiñ_alignments)
```

Creation of the dtw distance evaluation function, where we compare each x and y coordinate of the score/performance alignment for each feature and obtain a numebr which represents the total distance divided by the length of the alignment.

This evaluation metric is saved in a dataframe one containing all the chroma distances and the other one all the spectra distances.

```python
#dtw_distance
def dtw_distance(s1, s2):
    # calcula la distancia DTW
    dtw_distance = 0
    for i in range(len(s1)):
      dtw_distance += abs(s1[i] - s2[i])
    return dtw_distance/len(s1)

dist_chroma = pd.DataFrame(columns=['txt','distance_chroma'])
dist_spectra = pd.DataFrame(columns=['txt','distance_spectra'])

for alignments in list_muiñ_alignments:
    files = os.listdir(alignments)
    txts=[]
    txts.append(os.path.join(alignments,files[0]))
    txts.append(os.path.join(alignments,files[1]))
    for txt in txts:
        if 'chroma' in txt:
            df = pd.read_csv(txt, sep = ",", names=['score_time', 'perf_time'], header=0)
            distance = dtw_distance(df["score_time"], df["perf_time"])
            dist_chroma = dist_chroma.append({'txt': alignments, 'distance_chroma': distance}, ignore_index=True)
        if 'spectra' in txt:
            df = pd.read_csv(txt, sep = ",", names=['score_time', 'perf_time'], header=0)
            distance = dtw_distance(np.array(df["score_time"]), np.array(df["perf_time"]))
            dist_spectra = dist_spectra.append({'txt': alignments, 'distance_spectra': distance}, ignore_index=True)
```

```python
display(dist_spectra)
```

```python
display(dist_chroma)
```

Finally, we merge these evaluation metrics with all the previous results obtained.

```python
dist_merged = dist_chroma.merge(dist_spectra, on='txt')
dist_merged
```

```python
dist_merged['distance_total'] =0
for i in range(len(dist_merged)):
      if dist_merged['distance_chroma'][i] < dist_merged['distance_spectra'][i]:
        dist_merged['distance_total'][i] = 'Chroma'
      else:
        dist_merged['distance_total'][i] = 'Spectra'

display(dist_merged)
```

```python
num_chroma = dist_merged['distance_total'].value_counts()['Chroma']
print('Chroma percentage',(num_chroma/len(dist_merged))*100,'%')

num_spectra = dist_merged['distance_total'].value_counts()['Spectra']
print('Spectra percentage',(num_spectra/len(dist_merged))*100,'%')

if num_spectra>num_chroma:
  print("Spectra's performance is better for this dataset")
else:
  print("\nChroma's performance is better for this dataset")
```
